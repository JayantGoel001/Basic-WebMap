```python
import numpy as np
import pandas as pd
import folium as fo
```

```python
bMap = fo.Map(width='100%',height='100%',min_zoom=2)
```

```python
bMap
```

# Plotting Coordinates of Volcano

```python
x = fo.FeatureGroup(name='My Map')
```

```python
df = pd.read_csv("239.volcano.csv")
df.head()
```

```python
lat = df["Latitude"].values
long = df["Longitude"].values

Name = df["Name"].values
Country = df["Country"].values
total_dealth = df["TOTAL_DEATHS"].values
```

```python
for i in range(len(df)):
    if str(total_dealth[i]) !='nan':
        x.add_child(fo.Marker(location=[lat[i],long[i]],popup=Name[i]+",\n"+Country[i],icon=fo.Icon(color='darkred')))
    else:
        x.add_child(fo.Marker(location=[lat[i],long[i]],popup=Name[i]+",\n"+Country[i],icon=fo.Icon(color='green')))
```

```python
bMap.add_child(x)
```
