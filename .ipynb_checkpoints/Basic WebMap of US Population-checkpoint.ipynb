```python
import numpy as np
import pandas as pd
import folium as fo
```

```python
bMap = fo.Map(width='100%',height='100%',min_zoom=2)
```

```python
bMap
```

```python
fg = fo.FeatureGroup(name="My Map")
```

```python
pop = pd.read_csv("241.us+cities+pop.csv")

pop.head()
```

```python
lat = pop["lat"].values
long = pop["lon"].values
name = pop["name"].values
popu = pop["pop"].values
```

```python
for i in range(len(pop)):
    fg.add_child(fo.Marker(location=[lat[i],long[i]],popup=name[i]))
```

```python
bMap.add_child(fg)
```
